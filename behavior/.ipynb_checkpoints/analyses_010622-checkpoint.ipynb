```python
def run_gee(outcome, predictors, subj_covariates, char_covariates, interactions, df, cov_struct=sm.cov_struct.Independence()):
    
    # outcome + subject controls
    outcome_df = pd.melt(df, id_vars=['sub_id'] + subj_covariates, value_vars=outcome)
    del outcome_df['variable']
    outcome_df.rename(columns={'value':'outcome'}, inplace=True)

    # predictors of interest
    pred_dfs = []
    for p,pred in enumerate(predictors):
        pred_df = pd.melt(df, id_vars=['sub_id'], value_vars=pred)
        if p == 0:
            pred_df['role'] = pred_df['variable']
        del pred_df['variable']
        del pred_df['sub_id']
        pred_df.rename(columns={'value':'pred' + str(p+1)}, inplace=True)
        pred_dfs.append(pred_df)
    pred_df = pd.concat(pred_dfs, axis=1)
    
    gee_df = pd.concat([outcome_df, pred_df],axis=1)
    
    # character controls
    if char_covariates is not None:
        cov_dfs = []
        for c,cov in enumerate(char_covariates):
            cov_df = pd.melt(df, id_vars=['sub_id'], value_vars=cov)
            del cov_df['variable']
            del cov_df['sub_id']
            cov_df.rename(columns={'value':'cov' + str(c+1)}, inplace=True)
            cov_dfs.append(cov_df)
        cov_df = pd.concat(cov_dfs, axis=1)
        gee_df = pd.concat([gee_df, cov_df],axis=1)

    ## run GEE
    pred_ = (' + ').join(['pred' + str(c+1) for c in np.arange(0,len(predictors))])
    formula = 'outcome ~ role + ' + pred_ + ' + ' + (' + ').join(subj_covariates)
    if char_covariates is not None:
        cov_ = (' + ').join(['cov' + str(c+1) for c in np.arange(0,len(char_covariates))])
        formula = formula + ' + ' + cov_
    if interactions is not None:
        formula = formula + ' + ' + (' + ').join(interactions)

    gee = sm.GEE.from_formula(formula, groups="sub_id", data=gee_df, missing='drop',
                              cov_struct=cov_struct, family=sm.families.Gaussian()) 
    gee_res = gee.fit(maxiter=50) 
    return gee_res
```

### Behavior-Dots difference

```python
from scipy.spatial import distance
ego_pos = [1,0] # egocentric position
dist_change = []

mask = (np.isfinite(df['dots_affil.norm.first'])) & (np.isfinite(df['power.tendency.by_char_01']))
for s, sub in df[mask].iterrows():
    sub_dists = []
    for c, char in enumerate(character_roles):
        task_pos = [sub['affil.tendency.by_char_0' + str(c+1)], sub['power.tendency.by_char_0'  + str(c+1)]]
        dots_pos = [sub['dots_affil.norm.' + char], sub['dots_power.norm.' + char]]
        dots_dist = distance.euclidean(dots_pos, ego_pos)
        task_dist = distance.euclidean(task_pos, ego_pos) 
        dots_task_dist = distance.euclidean(dots_pos, task_pos)
        sub_dists.append(dots_task_dist)
    dist_change.append(sub_dists)
    
dist_df = pd.DataFrame(dist_change, columns=['distance_change.' + char for char in character_roles])
dist_df['distance_change.mean'] = np.mean(dist_df[['distance_change.' + char for char in character_roles]], axis=1)
dist_df = pd.concat([df[mask].reset_index(drop=True), dist_df], axis=1)
```

```python
melted = pd.melt(dist_df, id_vars=['sub_id', 'dx'], value_vars=['distance_change.' + char for char in character_roles + ['mean']])

fig, ax = plt.subplots(figsize=(5,5))
sns.barplot(x='dx', y='value', hue='variable', data=melted, edgecolor='.2')
sns.stripplot(x='dx', y='value', hue='variable', data=melted, 
             dodge=True, edgecolor=".2", alpha=.6, linewidth=1)
ax.set_title('Change in egocentric distance')
# ax.set_xticklabels(labels=['RES', 'PTSD', 'HC'], rotation=75)
handles, labels = plt.gca().get_legend_handles_labels()
l = plt.legend(title='', loc='upper right', handles=handles[6:],
               frameon=False, bbox_to_anchor=(1.6, 1), borderaxespad=0)
plt.show()
```

### Self-rank

```python
from scipy.stats import rankdata
def rankdata_rev(array, method='average'):
    return array.shape[1] - rankdata(array, method=method, axis=1) + 1
df['self_power'] = 0
ranked_power = rankdata_rev(df[['dots_power.norm.' + char for char in character_roles] + ['self_power']])
df['self_power.rank'] = ranked_power[:,5]
```

```python
long = pd.melt(df, id_vars=['sub_id', 'dx', 'sex', 'race___white', 'ctq_total_score_2'], 
               value_vars=['self_power.rank'])
model = ols("value ~ C(sex, Sum) + C(race___white, Sum) + C(ctq_total_score_2, Sum) + C(dx, Sum)", data=long).fit()
aov_table = sm.stats.anova_lm(model, typ=3)
display(aov_table)
```
