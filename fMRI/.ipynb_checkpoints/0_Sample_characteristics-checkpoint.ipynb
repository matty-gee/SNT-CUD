```python
def mahalanobis(x=None, data=None, cov=None):

    x_mu = x - np.mean(data)
    if not cov:
        cov = np.cov(data.values.T)
    inv_covmat = np.linalg.inv(cov)
    left = np.dot(x_mu, inv_covmat)
    mahal = np.dot(left, x_mu.T)
    return mahal.diagonal()

# mahalanobis(x=data[['sex','edu','age']], data=data[['sex','edu','age']])
# from scipy.stats import chi2

# #calculate p-value for each mahalanobis distance 
# df['p'] = 1 - chi2.cdf(df['mahalanobis'], 3)
```

```python
for demo_var in ['edu']:

    mask = data_mj[demo_var] != 99
    mask = data_mj[demo_var] != 999
    data_mj_ = data_mj[mask]
    data_mj_.reset_index(inplace=True, drop=True)
    
    mask = data_hc[demo_var] != 99
    mask = data_hc[demo_var] != 999
    data_hc_ = data_hc[mask]
    data_hc_.reset_index(inplace=True, drop=True)

    accs = []
    aucs = []
    tvals = []
    pvals = []
    n = 0
    
    while n < 1000:

        mj_ = np.random.choice(np.arange(1,len(data_mj_)), 100, replace=False)
        hc_ = np.random.choice(np.arange(1,len(data_hc_)), 100, replace=False)
        t, p = scipy.stats.ttest_ind(finite_mask(data_mj_[demo_var][mj_]), 
                                     finite_mask(data_hc_[demo_var][hc_]))
        
#         mj_m = np.sum(finite_mask(data_mj_[demo_var][mj_]) == 1)
#         mj_f = np.sum(finite_mask(data_mj_[demo_var][mj_]) == 2)
#         hc_m = np.sum(finite_mask(data_hc_[demo_var][hc_]) == 1)
#         hc_f = np.sum(finite_mask(data_hc_[demo_var][hc_]) == 2)
#         gender = [[mj_m,mj_f],[hc_m,hc_f]]
#         stat, p, dof, expected = chi2_contingency(gender)

        if p > 0.05:

            print(n)

            tvals.append(t)
            pvals.append(p)
            
            sub_ids_  = np.array(list(data_mj_['subject'][mj_]) + list(data_hc_['subject'][hc_]))
            incl_mask_ = pd.Series(sub_list).isin(sub_ids_).values

            X_ = X[incl_mask_]
            y_ = y[incl_mask_]
            training_ = training[incl_mask_]
            testing_ = testing[incl_mask_]

            X_train = X_[training_]
            X_test = X_[testing_]
            y_train = y_[training_]
            y_test = y_[testing_]

            clf = SGDClassifier(loss='hinge', penalty='l1', alpha=0.001, random_state=1, max_iter=10000)
            clf.fit(X_train, y_train)
            accs.append(clf.score(X_test, y_test))

            # for probabiltiies
            clf = CalibratedClassifierCV(clf, cv=5, method='sigmoid', ensemble=False)
            clf.fit(X_train, y_train)
            fpr, tpr, thresholds = roc_curve(y_test, clf.predict_proba(X_test)[:,1], pos_label=1)
            aucs.append(auc(fpr, tpr))

            n = n + 1

    equal_split = {'accs': accs, 'aucs': aucs, 'tvals': tvals, 'pvals': pvals}
    pickle_file(equal_split, demo_var + '_balanced_n200.pkl')
```

```python
def get_perm_pval(coef, perm_coefs, test='2-sided'):
    '''2-sided, greater or lesser
    '''
    p_greater = np.sum(perm_coefs >= coef) / len(perm_coefs)
    p_less  = np.sum(perm_coefs <= coef) / len(perm_coefs)
    p_2sided  = min(p_greater, p_less)
    if test == '2-sided':
        pval = p_2sided
    elif test == 'greater':
        pval = p_greater
    elif test == 'less':
        pval = p_less
    return pval
```

# Fit models to predict education high v low split

```python
def finite_mask(df, variable):
    return pd.DataFrame(df[np.isfinite(df[variable])][['sub_id', variable]])

def get_median_split(df, variable):
    df = finite_mask(df, variable)
    median_mask = df[variable] > np.median(df[variable])
    df.reset_index(inplace=True, drop=True)
    df.insert(2, 'median_split', median_mask*1)
    return df

def find_sub_ixs(new_sub_list):
    return np.array([i for i, e in enumerate(sub_list) if e in set(new_sub_list)])
```
