```python
fnames = glob.glob(f'{lsa_dir}/roi_timeseries/*.xlsx')
print(f'found {len(fnames)} files')

# get labels
fc_labels = pd.read_excel(fnames[0])['roi']
fc_labels = [l.translate(str.maketrans('', '', ''.join([' ', ',', '(', ')', "'", '-', '/']))) for l in fc_labels]
fc_label_cols = []
fc_label_vec  = []
for li1 in range(0, len(fc_labels) + 1):
    for li2 in range(li1 + 1, len(fc_labels)):
        fc_label_cols.append([fc_labels[li1], fc_labels[li2]])
        fc_label_vec.append(fc_labels[li1] + '_and_' + fc_labels[li2])
        
# get similarity/correlation (if pearsons, fisher z-transform)
get_corr = ConnectivityMeasure(kind='correlation', vectorize=True, discard_diagonal=True)
dfs      = [pd.DataFrame(np.arctanh(get_corr.fit_transform([pd.read_excel(f).iloc[:, 1:].values.T])[0])).T.astype(float) for f in fnames]

# make dataframe
corr_df = pd.concat(dfs)
corr_df.columns = fc_label_vec
corr_df.insert(0, 'sub_id', [int(f.split('/')[-1].split('_')[0]) for f in fnames])
corr_df.insert(1, 'dx',     [df[df['sub_id']==int(sub_id)]['dx'].values[0] for sub_id in corr_df['sub_id']])
# corr_df.to_excel(corr_fname, index=False)

# filter for inclusion
sub_list = df[df['memory_mean'] > .2]['sub_id'].values 
corr_df  = corr_df[corr_df['sub_id'].isin(sub_list)] 
corr_df  = corr_df.merge(df, on=['sub_id', 'dx']) # merge w/ all data
corr_df.head(5)
```

## Differences between groups

```python
hc_corrs = corr_df[corr_df['dx'] == 'HC'][fc_label_vec]
cd_corrs = corr_df[corr_df['dx'] == 'CD'][fc_label_vec]
print(f'CD n={len(cd_corrs)}, HC n={len(cd_corrs)}')

hcs_ = ut_vec_to_symm_mat(np.mean(hc_corrs, 0))
cds_ = ut_vec_to_symm_mat(np.mean(cd_corrs, 0))
diff = hcs_ - cds_

max_diff = np.round(np.max(np.abs(diff)),3)
print(f'max r_z diff={max_diff}') # max diff

# plot correlation matrix
plotting.plot_matrix(diff, labels=fc_labels, figure=(20, 17), 
                     vmin=-np.round(max_diff, 1), vmax=np.round(max_diff, 1))
plt.show()

## plot on the brain
## plotting.plot_connectome(mean_correlations_adhd, coords,
##                          edge_threshold="80%", title='Correlations')

# corr_df = corr_df.merge(df, on=['sub_id', 'dx'])
```

```python
apriori_rois = ['LIFG_Javadi_radius8','RIFG_Javadi_radius8',
                'LHPC_Tavares_radius8','RPrecuneus_Tavares_radius8',
                'LHPC_avg_radius8',
                'LHPC_ant','LHPC_post','LHPC_mid',
                'RHPC_ant','RHPC_post','RHPC_mid']
apriori_corrs = []
for li1 in range(0, len(apriori_rois) + 1):
    for li2 in range(li1 + 1, len(apriori_rois)):
        apriori_corrs.append(apriori_rois[li1] + '_and_' + apriori_rois[li2])
```

```python
# plot correlations
hcs_ = ut_vec_to_symm_mat(np.mean(hc_corrs[apriori_corrs], 0))
cds_ = ut_vec_to_symm_mat(np.mean(cd_corrs[apriori_corrs], 0))
corr_diff = hcs_ - cds_
max_diff = np.round(np.max(np.abs(corr_diff)), 3)
print(f'max r_z diff. = {max_diff}')

# plot correlation matrix
plotting.plot_matrix(np.tril(corr_diff), labels=apriori_rois, vmax=.15, vmin=-.15, figure=(6, 6))
plt.show()

# t-tests
for fc_col in apriori_corrs:
    t,t_p = scipy.stats.ttest_ind(hc_corrs.loc[:, fc_col].values, cd_corrs.loc[:,fc_col].values)
    if t_p < 0.05:
        res, ols = run_ols(['ctq_total_score_2', 'dx'], fc_col, corr_df, covariates=['age_years', 'sex'])
        ols_p    = np.round(res['dx[T.HC]_pvalue'].values[0],4)
        ols_b    = np.round(res['dx[T.HC]_beta'].values[0],2)
        print(f'{fc_col}: p={ols_p}, beta={ols_b}')
```

```python
sr_corr_df = pd.DataFrame(columns=['n', 'fc_region1', 'fc_region2', 'self-report', 'pvalue', 'beta'])
for sr_col in ['ctq_total_score_2', 
               'sni_num_ppl', 'cssa_total_mri',
               'lsas_social_interaction_fear_subscale', 
               'coc_age_1st_use', 'coc_days_last_use', 'mri_utox___coc']:
        
    X_labels = [sr_col]
                
    if 'coc' not in sr_col: # control for diagnosis if it's not already a cocaine variable
        X_labels.append('C(dx)')
        X_labels.append('C(dx)*' + sr_col)
        mask = np.isfinite(corr_df_z[sr_col])
    else: # ensure its only CUDs
        mask = np.isfinite(corr_df_z[sr_col]) & (corr_df_z['dx'] == 'CD')
    n = np.sum(mask)
                   
    for fc_col in label_cols:
        
        # should also filter for the other region too
        if 'Hippocampus' in fc_col or 'Precun' in fc_col or 'Cingulate' in fc_col: # simplify
            
            # to speed up, filter on bivariate correlation
            _, cor_p = scipy.stats.pearsonr(corr_df_z[fc_col][mask], corr_df_z[sr_col][mask])
            if cor_p < 0.05: 

                # add any interactions
                res_df, ols = run_ols(X_labels, fc_col, corr_df_z[mask], covariates=['age_years', 'sex', 'asi_education'])
                ols_p       = res_df[sr_col + '_pvalue'].values[0]
                ols_b       = res_df[sr_col + '_beta'].values[0]
                if ols_p < 0.005:
                    sr_corr_df.loc[len(sr_corr_df) + 1, :] = [n, fc_col.split('_')[0], fc_col.split('_')[1], sr_col, ols_p, ols_b]
sr_corr_df
```
