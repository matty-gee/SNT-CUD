### Compute weighted functional connectivity

```python
run_wc = dict()
for name, model in models.items():
    weights = model.coef_[0]
    wc = np.zeros((fcs.shape[0], vector_size))
    for i, fc in enumerate(fcs):
        wc[i,:] = np.abs(np.multiply(weights, fc)) 
    run_wc[name] = wc
    
sub_wc = []
for s in range(290):
    sub_wc.append(np.mean((run_wc[model_name][2*s,:], run_wc[model_name][2*s+1,:]), axis=0))
sub_wc = {model_name: np.array(sub_wc)}
```

## Regional degree centrality (DC)

```python
# Selected indices are the top 20 and bottom 2
selected_ind = np.hstack([np.arange(2), np.arange(70,90)])
x = np.arange(len(selected_ind))

for ix, (name, graph_list) in enumerate(all_graphs.items()): 

    print(name)
    # regional DC
    model_dc = dict()
    for graphobj_ in graph_list:
        graphobj = graphobj_[1]
        for parcel_name, properties in graphobj.graph.nodes.items():
            if parcel_name not in model_dc:
                model_dc[parcel_name] = [properties['degree_centrality']]
            else:
                model_dc[parcel_name].append(properties['degree_centrality'])

    # Calculate the mean and std of degree centrality for each parcel
    # Sort by highest to lowest and selected only the top 10, and bottom 2 for comparison
    avg_parcel_dc = np.zeros(num_parcels)
    std_parcel_dc = np.zeros(num_parcels)
    for i, (parcel_name, dcs) in enumerate(model_dc.items()):   
        avg_parcel_dc[i] = np.mean(dcs)
        std_parcel_dc[i] = np.std(dcs)
    sorted_ind = np.argsort(avg_parcel_dc)
    sortselect_ind = sorted_ind[selected_ind]

    if name == model_name: 
        
        print(parcel_names[sorted_ind])
        
        fig = go.Figure(layout=dict(
            height=500, width=800,
            template='ggplot2',
            title=f'{model_name}',
            xaxis=dict(
                title=dict(text='<b>Brain Region</b>'),
                tickvals=np.arange(len(selected_ind)),
                ticktext=parcel_descrip[sorted_ind][selected_ind],
                tickangle=45
            ),
            yaxis=dict(title=dict(text='<b>Degree Centrality</b>'))
        ))

        # Plot violin plot for each selected parcel and color
        sortselect_parcels = parcel_names[sortselect_ind]
        for parcel_name, dcs in model_dc.items():
            if parcel_name in sortselect_parcels:
                plot_ind = np.where(sortselect_parcels==parcel_name)[0][0]
                fig.add_trace(go.Violin(
                    x=np.repeat([plot_ind], len(dcs)), y=dcs, 
                    name=parcel_name, line_color=network_rgba[sortselect_ind][plot_ind],
                    box_visible=True, showlegend=False
                ))
        fig.show()
        fig.write_image(f'{network_fig_dir}/region_dc.svg')
```

```python
# Store average DC scores per parcel
model_avg_dc = dict()
for ix, (name, graph_list) in enumerate(all_graphs.items()):
    this_model_avg_dc = dict()
    for parc, dc_score in zip(parcel_names, avg_parcel_dc):
        this_model_avg_dc[parc] = dc_score
    model_avg_dc[name] = this_model_avg_dc
```

## Community detection

1. average weighted connectivity values across all subjects (note that these are absolute values)
2. threshold fcs at the community threshold
3. generate a GraphCorr object from matrix

```python
avg_thr_wc = dict()
for name, wc in sub_wc.items():
    avg_wc = np.mean(wc, 0)
    thr_wc = (avg_wc > np.percentile(avg_wc, comm_threshold)) * avg_wc
    # thr_wc = (avg_wc > comm_threshold) * avg_wc
    avg_thr_wc[name] = {'graphobj': GraphFromCorrv2(thr_wc, parcel_names, network_inds), 'orig_avg_wc': avg_wc}
```

```python
# Loop over all average GraphCorr objects to find communities
for ix, (name, thr_wc) in enumerate(avg_thr_wc.items()):
    
    g = thr_wc['graphobj']
    orig_avg_wc = thr_wc['orig_avg_wc']
    comms = thr_wc['graphobj'].find_communities(levels=7) # levels?
    avg_thr_wc[name]['communities'] = comms
    avg_thr_wc[name]['comm_inds'] = []

    # reorder the stanford indices in the order of communities
    reorg_comm_inds = np.array([])
    for comm in comms:
        avg_thr_wc[name]['comm_inds'].append(parcel_names_to_ind(comm, parcel_names))
        reorg_comm_inds = np.append(reorg_comm_inds, parcel_names_to_ind(comm, parcel_names))
    reorg_comm_inds = reorg_comm_inds.astype(int)
        
    if name == model_name:
        
        print(len(comms), 'communities detected')
        
        # plot by communities
        fig = go.Figure(layout=dict(
            title='Community averaged weighted connectivity',
            height=600, width=600,
            template='none',
            xaxis=dict(title=dict(text='<b>Region</b>')),
            yaxis=dict(title=dict(text='<b>Region</b>'))
        ))
        fig.add_trace(go.Heatmap(
            z=convert_upper_mat(orig_avg_wc)[reorg_comm_inds][:,reorg_comm_inds],
            colorscale='Reds'
        ))
        fig.show()
        # fig.write_image(f'{network_fig_dir}/{model_name}_community_sorted_wc.svg')
```

```python
# For each model find the average DC score for each community and order highest to lowest
for name, obj in avg_thr_wc.items():
    community_mean_dc = []
    for community in obj['communities']:
        scores = []
        for parcel in community:
            scores.append(model_avg_dc[name][parcel])
        community_mean_dc.append((np.round(np.mean(scores), 5), np.array(list(community))))
    avg_thr_wc[name]['comm_df'] = pd.DataFrame(community_mean_dc, 
                                               columns=['Mean DC', 'Community']).sort_values(by='Mean DC', ascending=False)
pickle_file(avg_thr_wc, f'{network_dir}/Network_analyses-site_demeaned.pkl')
```

```python
###########################
# get functional connectivities

infile = np.load(f'{output_dir}/functional_connectivity.npz')

run_y = infile['labels'][include_runs]
run_X = infile['fcs'][include_runs]
run_training = np.repeat(tt_df['training'][include_subs] == 1, 2)
run_testing  = np.repeat(tt_df['training'][include_subs] == 0, 2)

site_demeaned = True
if site_demeaned:
    # might want to only demean with the training average
    for runs_ in [mrn_runs, meg_runs]:
        run_X[runs_] = run_X[runs_] - np.mean(run_X[runs_ & training_runs]) # training mean 

X = []
for s in range(int(run_X.shape[0]/2)):
    X.append(np.mean((run_X[2*s,:], run_X[2*s+1,:]), axis=0))
X = np.array(X)

y = run_y[0::2]
testing = run_testing[0::2]
training = run_training[0::2]

###########################
# get trained model 

model_name = 'L1_LinearSVC'
with open(decoding_dir + '/Model_weights-site_demeaned.pkl', 'rb') as f:
    models = pickle.load(f)
model     = models[model_name]
weights   = model.coef_[0]
intercept = model.intercept_ 

# whole model/brain accuracy
whole_brain_acc = np.round(accuracy_score(run_y[run_testing], model.predict(run_X[run_testing])), 4)
chance_level = np.round(np.sum(run_y[run_testing])/len(run_y[run_testing]), 4) # will be 1-this when theres more HCs
print(whole_brain_acc * 100, 'accuracy vs. chance:', chance_level*100)


###########################
# get network analyses
stanford_coords = pd.read_csv(f'{data_dir}/stanford_coord_mapping.tsv', sep='\t')

with open(f'{network_dir}/Network_analyses-site_demeaned.pkl', 'rb') as f:
    network_analyses = pickle.load(f)
n_comms = len(network_analyses[model_name]['comm_inds']) 
comm_ixs = []
for c, (_, row) in enumerate(network_analyses[model_name]['comm_df'].iterrows()):
    comm_ixs.append([get_ind(stanford_coords, elem) for elem in row['Community']])
```

```python
run_dists = np.array([np.dot(fc, weights) + intercept for fc in run_X])
sub_dists = np.mean((run_dists[0::2], run_dists[1::2]), axis=0)

preds  = ((sub_dists > 0) * 1)[testing]
acc = np.round(accuracy_score(y[testing], preds),4)*100
print('subject level test predictions', acc)
```

```python
outfile  = f'{network_dir}/{model_name}_cumulative-community_decoding_training&testing-site_demeaned.xlsx'

# if os.path.isfile(outfile):
#     df = pd.read_excel(outfile)
# else:

# track accuracies
acc_0_thresh_train    = np.zeros(n_comms)
acc_mean_thresh_train = np.zeros(n_comms)
acc_dist_thresh_train = np.zeros(n_comms)
acc_perc_thresh_train = np.zeros(n_comms)

acc_0_thresh_test    = np.zeros(n_comms)
acc_mean_thresh_test = np.zeros(n_comms)
acc_dist_thresh_test = np.zeros(n_comms)
acc_perc_thresh_test = np.zeros(n_comms)

n_regions = []
n_corrs   = []
for c in range(len(comm_ixs)):

    # get the community indices, then the indices for alt the correlations
    region_ixs = get_region_ixs(comm_ixs[0:c+1])
    n_regions.append(len(region_ixs))
    corr_ixs = get_correlation_ixs(region_ixs)
    n_corrs.append(len(corr_ixs))

    train_dists = get_distances(X[training][:,corr_ixs], weights[corr_ixs], intercept)  
    test_dists  = get_distances(X[testing][:,corr_ixs], weights[corr_ixs], intercept)   

    # get thresholds
    top_performing = find_optimal_clf_thresh(train_dists, y[training])

    # predictions 
    dist_thresh = top_performing.iloc[0,1]
    perc_thresh = np.percentile(test_dists, top_performing.iloc[0,0])

    # sns.scatterplot(x='threshold', y='accuracy', data=top_performing)
    # plt.title(c+1)
    # plt.xlim(-100,100)
    # plt.show()

    acc_0_thresh_train[c]    = accuracy_score(y[training], (train_dists > 0).astype(int)) # 0 threshold
    acc_mean_thresh_train[c] = accuracy_score(y[training], (train_dists > np.mean(train_dists)).astype(int)) # mean threshold
    acc_dist_thresh_train[c] = accuracy_score(y[training], (train_dists > dist_thresh).astype(int)) # distance
    acc_perc_thresh_train[c] = accuracy_score(y[training], (train_dists > perc_thresh).astype(int)) # percentile

    acc_0_thresh_test[c]    = accuracy_score(y[testing], (test_dists > 0).astype(int)) # 0 threshold
    acc_mean_thresh_test[c] = accuracy_score(y[testing], (test_dists > np.mean(test_dists)).astype(int)) # mean threshold
    acc_dist_thresh_test[c] = accuracy_score(y[testing], (test_dists > dist_thresh).astype(int)) # distance
    acc_perc_thresh_test[c] = accuracy_score(y[testing], (test_dists > perc_thresh).astype(int)) # percentile

data = np.concatenate((np.array(n_regions).reshape(-1,1),
                       np.array(n_corrs).reshape(-1,1), 
                       acc_0_thresh_train.reshape(-1,1), acc_0_thresh_test.reshape(-1,1), 
                       acc_mean_thresh_train.reshape(-1,1), acc_mean_thresh_test.reshape(-1,1), 
                       acc_dist_thresh_train.reshape(-1,1), acc_dist_thresh_test.reshape(-1,1), 
                       acc_perc_thresh_train.reshape(-1,1), acc_perc_thresh_test.reshape(-1,1)), 
                       axis=1)
df = pd.DataFrame(data, columns=['num_regions','num_corrs', 
                                 'accuracy_0_train','accuracy_0_test', 
                                 'accuracy_mean_dist_train','accuracy_mean_dist_test', 
                                 'accuracy_best_dist_train','accuracy_best_dist_test', 
                                 'accuracy_best_percentile_train','accuracy_best_percentile_test'])
# df.to_excel(outfile, index=False)
```
