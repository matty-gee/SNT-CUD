## Compute correlations

```python
fnames   = glob.glob(f'{lsa_dir}/roi_timeseries/*.xlsx')
print(f'found {len(fnames)} files')

# get labels
fc_labels = pd.read_excel(fnames[0])['roi']
fc_labels = [l.translate(str.maketrans('', '', ''.join([' ', ',', '(', ')', "'", '-', '/']))) for l in fc_labels]
fc_label_cols = []
fc_label_vec  = []
for li1 in range(0, len(fc_labels) + 1):
    for li2 in range(li1 + 1, len(fc_labels)):
        fc_label_cols.append([fc_labels[li1], fc_labels[li2]])
        fc_label_vec.append(fc_labels[li1] + '_and_' + fc_labels[li2])
        
# get similarity/correlation (if pearsons, fisher z-transform)
get_corr = ConnectivityMeasure(kind='correlation', vectorize=True, discard_diagonal=True)
dfs      = [pd.DataFrame(get_corr.fit_transform([pd.read_excel(f).iloc[:, 1:].values.T])[0]).T.astype(float) for f in fnames]

# df
corr_df = pd.concat(dfs)
corr_df.columns = corr_label_vec
corr_df.insert(0, 'sub_id', [int(f.split('/')[-1].split('_')[0]) for f in fnames])
corr_df.insert(1, 'dx',     [df[df['sub_id']==int(sub_id)]['dx'].values[0] for sub_id in corr_df['sub_id']])
sub_list = df[df['memory_mean'] > .2]['sub_id'].values # filter for inclusion
corr_df  = corr_df[corr_df['sub_id'].isin(sub_list)] 
corr_df  = corr_df.merge(df, on=['sub_id', 'dx'])
# corr_df.to_excel(corr_fname, index=False)
corr_df.head(5)
```

#### Filter for possible interesting correlations
- focus on regions previously found: e.g., left hippocampus, precuneus/pcc

```python
sr_corr_df = pd.DataFrame(columns=['n', 'fc_region1', 'fc_region2', 'self-report', 'pvalue', 'beta'])
for sr_col in ['ctq_total_score_2', 
               'sni_num_ppl', 'cssa_total_mri',
               'lsas_social_interaction_fear_subscale', 
               'coc_age_1st_use', 'coc_days_last_use', 'mri_utox___coc']:
        
    X_labels = [sr_col]
                
    if 'coc' not in sr_col: # control for diagnosis if it's not already a cocaine variable
        X_labels.append('C(dx)')
        X_labels.append('C(dx)*' + sr_col)
        mask = np.isfinite(corr_df_z[sr_col])
    else: # ensure its only CUDs
        mask = np.isfinite(corr_df_z[sr_col]) & (corr_df_z['dx'] == 'CD')
    n = np.sum(mask)
                   
    for fc_col in label_cols:
        
        # should also filter for the other region too
        if 'Hippocampus' in fc_col or 'Precun' in fc_col or 'Cingulate' in fc_col: # simplify
            
            # to speed up, filter on bivariate correlation
            _, cor_p = scipy.stats.pearsonr(corr_df_z[fc_col][mask], corr_df_z[sr_col][mask])
            if cor_p < 0.05: 

                # add any interactions
                res_df, ols = run_ols(X_labels, fc_col, corr_df_z[mask], covariates=['age_years', 'sex', 'asi_education'])
                ols_p       = res_df[sr_col + '_pvalue'].values[0]
                ols_b       = res_df[sr_col + '_beta'].values[0]
                if ols_p < 0.005:
                    sr_corr_df.loc[len(sr_corr_df) + 1, :] = [n, fc_col.split('_')[0], fc_col.split('_')[1], sr_col, ols_p, ols_b]
sr_corr_df
```
